```python
fig, axes = plt.subplots(2,1 , figsize=(10, 10))
Year = df.groupby("Year")["Year"].count()
sns.lineplot(data=Year,ax=axes[0],color="green")
axes[0].set_xlabel("Year of Release")
axes[0].set_ylabel("No of Movies Released")
axes[0].set_title("Graph showing how many movies are released each year (1917-Present(2025))")

genre_data = df.groupby(["Year", "Genre"])["Genre"].count().unstack()
for genre in ["Action", "Animation", "Biography", "Comedy", "Drama"]:
    sns.lineplot(data=genre_data[genre],label=genre)
axes[1].set_xlabel("Year of Release")
axes[1].set_ylabel("Number of Movies Released")
axes[1].set_title("Graph showing how many movies are released each year (1917-Present(2025)) (per Genre)")
axes[1].legend(title="Genres")
plt.tight_layout()
plt.show()
```

##### The First Graph shows the significant increase of films from 1980 to present . And the peak of these genres are noted in the year 2023
##### If we observe clearly we can also observe the fall in the number of films released at the time of pandemic
##### The Second Graph Shows that the increase in Biography was too high upto 2018-19 but after that people began to encourage Action and Animation. Comedy and Drame have been consistent since 1990s
